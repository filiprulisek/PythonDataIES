```python
import requests

# Define a list to store all matches
all_matches = []

# Loop through rounds 1 to 34
for round_number in range(1, 35):
    url = f"https://www.sofascore.com/api/v1/unique-tournament/35/season/52608/events/round/{round_number}"
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        matches = response.json()["events"]
        all_matches.append(matches)
    else:
        print(f"Failed to retrieve matches for round {round_number}")

# Print all matches
for idx, matches in enumerate(all_matches, start=1):
    print(f"Matches for round {idx}:")
    for match in matches:
        print(match)
    print("\n")
```

```python
def extract_events(game_data):
    events = []

    for incident in game_data['incidents']:
        if incident['incidentType'] == 'goal':
            goal_event = {
                'Event': 'Goal',
                'Minute': incident['time'],
                'Scorer': incident['player']['shortName'],
                'Assist': incident.get('assist1', {}).get('shortName', '-'),
                'Team': 'Home' if incident['isHome'] else 'Away'
            }
            events.append(goal_event)

        elif incident['incidentType'] == 'substitution':
            sub_event = {
                'Event': 'Substitution',
                'Minute': incident['time'],
                'Player In': incident['playerIn']['shortName'],
                'Player Out': incident['playerOut']['shortName'],
                'Team': 'Home' if incident['isHome'] else 'Away'
            }
            events.append(sub_event)

        elif incident['incidentType'] == 'card':
            card_event = {
                'Event': 'Card',
                'Minute': incident['time'],
                'Player': incident['player']['shortName'],
                'Reason': incident['reason'],
                'Card Type': incident['incidentClass'],  # 'yellow', 'red', etc.
                'Team': 'Home' if incident['isHome'] else 'Away'
            }
            events.append(card_event)

    return events
def create_dataframe(events):
    # Create separate dataframes for home and away events
    df_home = pd.DataFrame([event for event in events if event['Team'] == 'Home'])
    df_away = pd.DataFrame([event for event in events if event['Team'] == 'Away'])

    # Rename columns to reflect home and away differentiation
    df_home.columns = [f'Home {col}' if col != 'Minute' else col for col in df_home.columns]
    df_away.columns = [f'Away {col}' if col != 'Minute' else col for col in df_away.columns]

    # Merge home and away dataframes on the Minute column
    df = pd.merge(df_home, df_away, on='Minute', how='outer')

    return df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to get the squad data for a team
def get_team_squad(team_url):
    response = requests.get(team_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    squad_table = soup.find('table', {'class': 'items'})

    if squad_table:
        players = squad_table.find_all('tr', {'class': ['odd', 'even']})
        team_squad = []

        for player in players:
            columns = player.find_all('td')
            player_data = {
                'Team': soup.find('h1').text.strip(),
                'Player': columns[1].text.strip(),
                'Number': columns[2].text.strip(),
                'Age': columns[4].text.strip(),
                'Country': columns[5].img['alt'] if columns[5].img else ''
            }
            team_squad.append(player_data)

        return team_squad
    return []

# Base URL for Bundesliga teams on Transfermarkt
base_url = 'https://www.transfermarkt.com/1-bundesliga/startseite/wettbewerb/L1/plus/?saison_id=2023'

# Get the main page content
response = requests.get(base_url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all team URLs
teams_table = soup.find('table', {'class': 'items'})
team_links = teams_table.find_all('a', {'class': 'vereinprofil_tooltip'})

# List to hold all squads
all_squads = []

# Loop through each team link and get squad data
for link in team_links:
    team_url = 'https://www.transfermarkt.com' + link['href']
    team_squad = get_team_squad(team_url)
    all_squads.extend(team_squad)
    time.sleep(1)  # Pause to avoid overloading the server

# Create a DataFrame from the scraped data
df_squads = pd.DataFrame(all_squads)

# Display the DataFrame
print(df_squads)

# Optionally, save the DataFrame to a CSV file
df_squads.to_csv('bundesliga_23_24_squads.csv', index=False)
```
